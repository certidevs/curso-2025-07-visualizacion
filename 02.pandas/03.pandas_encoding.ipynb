## Encoding con .map() 

Interesa cuando las categorías tienen un orden lógico o jeráquico. Ejemplos:
    * bajo, medio, alto
    * colegio, instituto, bachiller, universidad, carrera, máster, doctorado

Ventaja: no crea columnas extra como el get_dummies, por lo que nos simplifica los cálculos en modelado.

Habitualmente se suele emplear en la variable de salida (y) en el modelado.

A mayores mostramos la correlación con df.corr()

```python
# cargar dataset iris
import pandas as pd

df = pd.read_csv('../Data/iris.csv')
df.head()
```

```python
df['Clase'].value_counts()
```

```python
# encoding usando .map() de texto a número 1, 2, 3, ....
# Ideal para situaciones donde la variable es ordinal, es decir, tiene un orden lógico
# como por ejemplo baja, media, alta 
mapeo_clase = {
    'Iris-setosa': 1,
    'Iris-versicolor': 2,
    'Iris-virginica': 3
}

df['Clase'] = df['Clase'].map(mapeo_clase)
df.head()
```

```python
# matriz de correlación
correlacion = df.corr()
correlacion
```

```python
# mostrar la matriz de correlación con seaborn:
import seaborn as sns

sns.heatmap(correlacion, annot=True, cmap='viridis')
```

## Encoding con .get_dummies()

En modelado, entrada X, salida y. 

Columnas One-hot. Variables nominales, (no ordinales). Tiene sentido cuando las categorías no tienen un orden intrínseco.

Análisis exploratorio de datos EDAs. Entender mejor las correlaciones al tener cada valor separado como una columna independiente. Mayor detalle por categoría para poder analizar las correlaciones.

Habitualmente se suele emplear en las variable de entrada (X) en el modelado.

```python
df = pd.read_csv('../Data/iris.csv')
```

```python
# genera una columna nueva por cada valor categórico y la rellena con 0 o 1
pd.get_dummies(df)
```

```python
# si solo queremos aplicar el encoding a columnas específicas podemos hacerlo así:
df_encoded = pd.get_dummies(df, columns=['Clase'])
df_encoded
```

```python
# importante: aplicar la correlación sobre el dataframe encoded no el original:
correlacion = df_encoded.corr()
sns.heatmap(correlacion, annot=True, cmap='viridis')
```
