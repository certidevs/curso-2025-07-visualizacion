```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ver datasets demo disponibles para pruebas con seaborn
sns.get_dataset_names()
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

## histplot

```python
# df.hist()
sns.histplot(df, x='fare');
```

```python
sns.histplot(df, x='fare', bins=20);
```

```python
sns.histplot(df, x='fare', bins=20, hue='alive');
```

## kdeplot

```python
sns.kdeplot(df, x='fare');
```

```python
sns.kdeplot(df, x='fare', hue='alive');
```

```python
sns.kdeplot(df[df['alive'] == 'yes'], x='fare');
sns.kdeplot(df[df['alive'] == 'no'], x='fare');
```

```python
sns.kdeplot(df, x='fare', y='age')
```

## countplot

Ideal para ver el conteo de categorías en una columna categórica.

Se usa mucho en problemas de machine learning aprendizaje supervisado de tipo clasificación para ver si las clases están balanceadas.

* Columnas numéricas: histplot, kdeplot
* Columnas categóricas: countplot

```python
sns.countplot(df, x='class');
```

## boxplot

resumen de estadísticas, ver medianas, cuartiles, outliers

```python
sns.boxplot(df, x='age');
```

```python
sns.boxplot(df, y='age');
```

```python
sns.boxplot(df, x='class', y='age');
```

```python
# sns.boxplot(df[df['fare'] < 150], x='survived', y='fare'); #aplicando filtro ya no haría falta zoom

sns.boxplot(df, x='survived', y='fare');
plt.ylim(-5, 130) # ZOOM PARA VER EL ÁREA INTERESANTE DONDE SE CONCENTRAN LOS DATOS
```

```python
sns.boxplot(df, x='survived', y='age');
```

## violinplot

boxplot + distribución 

cuando además del resumen de estadísticas quieres ver la distribución

```python
sns.violinplot(df, y='age')
```

## stripplot

puntos simples para ver dónde se concentran los datos. El problema es que los puntos se solapan.

Si hay muchos registros (filas) podría ser problemático pintar tantos puntos (costoso computacionalmente)

```python
sns.stripplot(df, y='age')
```

```python
sns.stripplot(df, x='survived', y='fare')
```

## swarmplot

puntos sin solapamiento, ideal si quieres los puntos individuales

MEjor en datasets pequeños ya que es más costoso de calcular, hay que tener cuidado.

Similar a violinplot pero viendo los puntos individuales distribuidos.

```python
sns.swarmplot(df, y='age')
```

```python
sns.violinplot(df, y='age', color='lightgray')
sns.swarmplot(df, y='age', size=4, color='black')
```
