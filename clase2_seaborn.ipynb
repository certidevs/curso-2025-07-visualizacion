```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ver datasets demo disponibles para pruebas con seaborn
sns.get_dataset_names()
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

## histplot

```python
# df.hist()
sns.histplot(df, x='fare');
```

```python
sns.histplot(df, x='fare', bins=20);
```

```python
sns.histplot(df, x='fare', bins=20, hue='alive');
```

## kdeplot

```python
sns.kdeplot(df, x='fare');
```

```python
sns.kdeplot(df, x='fare', hue='alive');
```

```python
sns.kdeplot(df[df['alive'] == 'yes'], x='fare');
sns.kdeplot(df[df['alive'] == 'no'], x='fare');
```

```python
sns.kdeplot(df, x='fare', y='age')
```

## countplot

Ideal para ver el conteo de categoras en una columna categrica.

Se usa mucho en problemas de machine learning aprendizaje supervisado de tipo clasificacin para ver si las clases estn balanceadas.

* Columnas numricas: histplot, kdeplot
* Columnas categricas: countplot

```python
sns.countplot(df, x='class');
```

## boxplot

resumen de estadsticas, ver medianas, cuartiles, outliers

```python
sns.boxplot(df, x='age');
```

```python
sns.boxplot(df, y='age');
```

```python
sns.boxplot(df, x='class', y='age');
```

```python
# sns.boxplot(df[df['fare'] < 150], x='survived', y='fare'); #aplicando filtro ya no hara falta zoom

sns.boxplot(df, x='survived', y='fare');
plt.ylim(-5, 130) # ZOOM PARA VER EL REA INTERESANTE DONDE SE CONCENTRAN LOS DATOS
```

```python
sns.boxplot(df, x='survived', y='age');
```

## violinplot

boxplot + distribucin 

cuando adems del resumen de estadsticas quieres ver la distribucin

```python
sns.violinplot(df, y='age')
```

## stripplot

puntos simples para ver dnde se concentran los datos. El problema es que los puntos se solapan.

Si hay muchos registros (filas) podra ser problemtico pintar tantos puntos (costoso computacionalmente)

```python
sns.stripplot(df, y='age')
```

```python
sns.stripplot(df, x='survived', y='fare')
```

## swarmplot

puntos sin solapamiento, ideal si quieres los puntos individuales

MEjor en datasets pequeos ya que es ms costoso de calcular, hay que tener cuidado.

Similar a violinplot pero viendo los puntos individuales distribuidos.

```python
sns.swarmplot(df, y='age')
```

```python
sns.violinplot(df, y='age', color='lightgray')
sns.swarmplot(df, y='age', size=4, color='black')
```

## scatterplot

ver la relacin entre dos variables numricas en un grfico de puntos

El parmetro hue ayuda a colorear los puntos en base a una variable categrica.

```python
sns.scatterplot(df, x='age', y = 'fare')
```

```python
sns.scatterplot(df, x='age', y = 'fare', hue='class')
plt.ylim(0, 100)
```

## heatmap

mapa de calor, para matrices o dataframes, por ejemplo para matriz de correlacin o tablas pivotantes.

Simplifica ms el proceso de visualizar las correlaciones con respecto a matplotlib.

```python
# correlacin
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='viridis')
```

```python
df_flights.head()
```

```python
# tabla pivotante
df_pivot = df_flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(df_pivot, annot=True, fmt="d", cmap='viridis')
```

```python
sns.clustermap(df_pivot, cmap='viridis', figsize=(8, 6))
```
